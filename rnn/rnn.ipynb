RNN with LSTM

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
# install Tensorflow and Keras in Anaconda
from keras.models import Sequential
from keras.layers import Dense, LSTM
from sklearn.metrics import accuracy_score
```

The data contains 14 channels and the last column is the state of the eye.

```python
eye_data = pd.read_csv('../eeg_eye.arff')
eye_data
```

```python
eye_data.describe()
```

```python
def normalization(x):
     return ((x - min(x)) / (max(x) - min(x)))
   
```

```python
def normData(data):
    for i in ['AF3','F7','F3','FC5','T7','P7','O1','O2','P8','T8','FC6','F4','F8','AF4', 'eye']:
        data[i] = normalization(data[i])
    return data
```

```python
normData(eye_data)
eye_data
```

```python
eye_data.describe()
```

```python
x = eye_data.drop(columns=['eye'])
y = eye_data['eye']

data=np.array(x, dtype=float)
target=np.array(y, dtype=float)
```

Custom data to learn RNN:

```python

# Divide the data into training data and test data  
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=4)

# RNN Model
model = Sequential()
# output size: 1
# number of inputs: None - if not known
# lenght of input sequences: 14 (5-for custom data)
# length of each vector: 1
model.add(LSTM((1), batch_input_shape=(None,14,1), return_sequences=True))
model.add(LSTM((1), return_sequences=False))
#binary_crossentropy
#mean_absolute_error
model.compile(optimizer='adam', loss='mean_absolute_error', metrics=['accuracy'])
model.summary()
```

```python
#custom - 400
history = model.fit(x_train, y_train, epochs=50, validation_data=(x_test,y_test))
```

```python
prediction = model.predict(x_test)
```

```python
import matplotlib.pyplot as plt
#for eye data - range(2996)
#for custom - 20
plt.scatter(range(2996),prediction,c='r')
plt.scatter(range(2996),y_test,c='g')
plt.show()
```
